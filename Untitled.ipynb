# Introduction
The goal of this project is to examine tweets relating to covid-19 in order to see what kinds of information is spreading in the world. What are the different feelings people have toward covid related issues? Once this becomes clear, we can identify misinformation and see where it's stemming from...

```python
import twint
import pandas as pd

import nest_asyncio # for some reason, needed to excute the api call
nest_asyncio.apply()
```

```python
c = twint.Config()
c.Search = "covid"
c.Min_likes = 100
c.Count = True
c.Limit = 10000
c.Store_csv = True
c.Output = 'covid_tweets.csv'
```

```python
twint.run.Search(c)
```

```python
tweets = pd.read_csv('covid_tweets.csv')
```

```python
tweets.head()
```

```python
tweets.drop(columns=['conversation_id', 'created_at', 'time', 'timezone', 'place', 'geo', 'source', 'user_rt_id', 'user_rt', 'retweet_id', 'reply_to', 'retweet_date', 'translate', 'trans_src', 'trans_dest','photos', 'quote_url', 'link', 'video', 'thumbnail', 'near'], axis=1, inplace=True)
```

```python
tweets.shape
```

```python
tweets.head()
```

```python
tweets = tweets[tweets['language'] == 'en']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_df=0.9, min_df=2, stop_words='english')
```

```python
dtm = cv.fit_transform(tweets['tweet'])
```

```python
dtm
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
LDA = LatentDirichletAllocation(n_components=7, random_state=42)
```

```python
LDA.fit(dtm)
```

```python
# Grab vocab of words
cv.get_feature_names()[5122]
```

```python
import random
random_word_id = random.randint(0,7859)

cv.get_feature_names()[random_word_id]
```

```python
# grab topics
single_topic = LDA.components_[0]
single_topic.argsort()
```

```python
single_topic.argsort()[-10:]
```

```python
top_fifteen = single_topic.argsort()[-20:]
```

```python
for index in top_fifteen:
    print(cv.get_feature_names()[index])
```

```python
# highest probablility topics
for index,topic in enumerate(LDA.components_):
    print(f"Top 20 words for topic #{index}")
    print([cv.get_feature_names()[index] for index in topic.argsort()[-20:]])
    print('\n')
    print('\n')
```

```python
topic_results = LDA.transform(dtm)
```

```python
topic_results[0].argmax()
```

```python
tweets['Topic'] = topic_results.argmax(axis=1)
```
